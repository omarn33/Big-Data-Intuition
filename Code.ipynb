```python
import pandas as pd

# Import data
data_2010 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2010.csv')
data_2011 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2011.csv')
data_2012 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2012.csv')
data_2013 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2013.csv')
data_2014 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2014.csv')
data_2015 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2015.csv')
data_2016 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2016.csv')
data_2017 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2017.csv')
data_2018 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2018.csv')
data_2019 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2019.csv')
data_2020 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2020.csv')
data_2021 = pd.read_csv('../input/elon-musk-tweets-2010-2021/2021.csv')
```

```python
# Merge all datasets
data = pd.concat([data_2010, data_2011, data_2012, data_2013, data_2014, data_2015, data_2016, data_2017, data_2018, data_2019, data_2020, data_2021], axis = 0)
```

```python
# Display table schema
pd.set_option("display.max_columns", 50)
data.head(5)
```

```python
# Question 1: Average Number of Likes Per Post

avg_likes_per_post = data['nlikes'].mean()

print(f"Average Number of Likes Per Post: {avg_likes_per_post}")
```

```python
# Question 2: Average Number of Views Per Post

# Create new column in the dataset to calculate the number of views per post
data['views'] = 20 * (data['nlikes'] + data['nreplies'] + data['nretweets'])
avg_views_per_post = data['views'].mean()

print(f"Average Number of Views Per Post: {avg_views_per_post}")
```

```python
# Question 3: Average Amount of Data (Bytes) Stored Per Post

# Average Byte Size of Photo
AVG_PHOTO_SIZE = 5000000

# Maximum Bytes Twitter allows per post
AVG_VIDEO_SIZE = 100000000
MAX_VIDEO_SIZE = 512000000

# Total number of photos tweeted
total_photos = len(data[data['photos'] != '[]']['photos'])

# Total number of videos tweeted
total_videos = len(data[data['video'] == 1]['video'])

# Total number of posts
total_posts = len(data['id'])

# Calculate the average data amount per post
avg_data_per_post = AVG_PHOTO_SIZE * (total_photos / total_posts) #+ AVG_VIDEO_SIZE * (total_videos / total_posts)

print(f"Average Amount of Data Stored Per Post: {avg_data_per_post} bytes")
```

```python
# Question 4: Total Data Transfer Per Post

# Calculate the total data transfer
total_data_transfer = avg_views_per_post * avg_views_per_post

print(f"Total Data Transfer Per Post: {total_data_transfer} bytes")
```

```python
# Question 5: Tweets Viewed Per Minute

# Total number of minutes since data collection started
TOTAL_MINS = 5785560

# Total number of views
total_views = data['views'].sum()

# Calculate the total tweets per minute
total_tweets_per_min = total_views / TOTAL_MINS

print(f"Tweets Viewed Per Minute: {total_tweets_per_min}")
```

```python
# Question 6: Data Accessed Per Minute

data_accessed_per_min = total_tweets_per_min * avg_data_per_post

print(f"Data Acccessed Per Minute: {data_accessed_per_min} bytes")
```

```python
# Question 7: 

# Function Selected: Updating Likes Counter on a Tweet
# Goal: Find the amount of Data Accessed when updating 'like' counter from 2019-2021

# Assumptions:
# - No Twitter users "undos" any likes
# - 95% of 'likes' happen within a month
# - When the # of likes is accessed by the operation, it is 3 bytes
# - Data touched when accessing the tweet id is negligible 

TOTAL_MONTHS = 36

# Recent posts include posts from 2019 to 2021
recent_total_posts = len(data_2019['id']) + len(data_2020['id']) + len(data_2021['id'])
avg_tweet_per_month = recent_total_posts / (TOTAL_MONTHS)
avg_likes_per_month = (0.95 * avg_likes_per_post) * avg_tweet_per_month

# 3 Bytes
SIZE_OF_LIKES_COUNTER_DATA = 3

data_accessed_per_month = SIZE_OF_LIKES_COUNTER_DATA * avg_likes_per_month

print(f"Average Amount of Data Accessed Per Month: {data_accessed_per_month} bytes")
```
